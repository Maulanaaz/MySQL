```python
# Merge tables with key value (INNER JOIN)
import mysql.connector
import pandas as pd
conn = mysql.connector.connect(host='localhost',
                              user='root',
                              passwd='***',
                              database='sakila')
query = """
SELECT * FROM actor, actor_info
WHERE actor.actor_id = actor_info.actor_id;
"""
df = pd.read_sql(query,conn)
df
```

```python
# Merge tables with key value (INNER JOIN)
import mysql.connector
import pandas as pd
conn = mysql.connector.connect(host='localhost',
                              user='root',
                              passwd='***',
                              database='sakila')
query = """
SELECT * FROM actor_info, actor
WHERE actor.actor_id = actor_info.actor_id;
"""
df = pd.read_sql(query,conn)
df
```

```python
# INNER JOIN with Prefix using condition
import mysql.connector
import pandas as pd
conn = mysql.connector.connect(host='localhost',
                              user='root',
                              passwd='***',
                              database='sakila')
query = """
SELECT actor.*,actor_info.*
FROM actor, actor_info
WHERE actor.actor_id = actor_info.actor_id;
"""
df = pd.read_sql(query,conn)
df
```

```python
# CROSS JOIN
import mysql.connector
import pandas as pd
conn = mysql.connector.connect(host='localhost',
                              user='root',
                              passwd='***',
                              database='sakila')
query = """
SELECT * FROM actor, actor_info;
"""
df = pd.read_sql(query,conn)
df
```

```python
# INNER JOIN (Intersection) Function
import mysql.connector
import pandas as pd
conn = mysql.connector.connect(host='localhost',
                              user='root',
                              passwd='***',
                              database='sakila')
query = """
SELECT * FROM actor
INNER JOIN actor_info
ON actor.actor_id = actor_info.actor_id;
"""
df = pd.read_sql(query,conn)
df
```

```python
# INNER JOIN
import mysql.connector
import pandas as pd
conn = mysql.connector.connect(host='localhost',
                              user='root',
                              passwd='***',
                              database='sakila')
query = """
SELECT title, release_year, description, language.name AS language
FROM film
INNER JOIN language
ON film.language_id = language.language_id;
"""
df = pd.read_sql(query,conn)
df
```

```python
# UNION Function with Condition
import mysql.connector
import pandas as pd
conn = mysql.connector.connect(host='localhost',
                              user='root',
                              passwd='***',
                              database='practice')
query = """
SELECT * FROM monday
WHERE attendance = "no"
UNION
SELECT * FROM tuesday
WHERE attendance = "no";
"""
df = pd.read_sql(query,conn)
df
```
